```python
orb = cv2.ORB_create()
```

```python
kp1, des1 = orb.detectAndCompute(img1, None)
```

```python
img2 = cv2.imread("images/2005.jpg")
plt.imshow(cv2.cvtColor(im2, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
kp2, des2 = orb.detectAndCompute(img2, None)
```

```python
# Create BFMatcher object
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)

# Match descriptors
matches = bf.match(des1, des2)

# Sort them in the order of their distance
matches = sorted(matches, key=lambda x: x.distance)
```

```python
# Draw first 10 matches
img3 = cv2.drawMatches(img1, kp1, img2, kp2, matches[:10], None, flags=cv2.DrawMatchesFlags_DRAW_RICH_KEYPOINTS)

plt.figure(figsize=(24, 24))
plt.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python

# Define the fixed size
fixed_size = (256, 256)
```

```python

img1_resized = cv2.resize(img1, fixed_size)
img2_resized = cv2.resize(img2, fixed_size)

# Convert to grayscale
gray1 = cv2.cvtColor(img1_resized, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2_resized, cv2.COLOR_BGR2GRAY)

hog = cv2.HOGDescriptor()
features1 = hog.compute(gray1)

features2 = hog.compute(gray2)
```

```python
from scipy.spatial import distance

import time

start_time = time.time()

# Your code here
dist = distance.euclidean(features1, features2)
print(f"Euclidean distance between images: {dist}")

end_time = time.time()
execution_time = end_time - start_time
print(f"Execution time: {execution_time} seconds")
```
